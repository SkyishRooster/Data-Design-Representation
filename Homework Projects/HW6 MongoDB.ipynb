### 1. (Let’s get things started …) Please write code (Python or Java) to query and print to screen all Pokémon character “name”s (and “_id” but not the entire document) with candy_count >= month + day of your birthday  (e.g., my birthday is 2/12 and I query candy_count >= 14 as 2+12 = 14).  (25% of points)   (Note:  the MongoDB operator for “>=” is “$gte”)

```python
#!pip install pymongo
```

```python
from pymongo import MongoClient
import regex as re
```

```python
# Connect to MongoDB
client = MongoClient('localhost', 27017)

# Access Database
db = client['samples_pokemon']

# Access Collection
collect = db['samples_pokemon']
```

```python
#print to screen all Pokémon character “name”s and “_id” with candy_count >= month + day of your birthday (e.g., my birthday is 2/12 and I query candy_count >= 14 as 2+12 = 14)
birthsum = 11 + 7
cursor = collect.find({"candy_count": {"$gte": birthsum}}, {"name":1})
for i in cursor:
    print(i)
```

### 2. (Let’s sprinkle in a little or …) Please write code (Python or Java) to query and print to screen all Pokémon character “name”s (and “_id” but not the entire document) with num = month or num = day of your birthday  (e.g., my birthday is 2/12 and I have to query num = 2 or num = 12).  (25% of points)

```python
#print to screen all Pokémon character “name”s and "_id" with num = month or num = day of your birthday 
cursor = collect.find({"num": {"$in": ["011", "007"]}}, {"name":1})
for i in cursor:
    print(i)
```

***MongoDB supports RegEx.  The MongoDB operator for RegEx is "\\$regex".  For example, to find all Crunchbase documents that have a (company) name that starts with a lower case character, we can search for "db.crunchbase_database.find({"name" : {\$regex : "^[a-z].\*"}})***

### 3. (And some RegEx as well …) Please write code (Python or Java) to query and print to screen all Crunchbase company “name”s (and “_id” but not the entire document) that have “text” in their “tag_list”.  (25% of points)

```python
#print to screen all Crunchbase company “name”s and “_id” that have “text” in their “tag_list”
db_crun = client['crunchbase']
collect_crun = db_crun['crunchbase_database']
cursor = collect_crun.find({"tag_list": {"$regex":".*text.*"}}, {"name":1})

for i in cursor:
    print(i)
```

### 4. (This is the final enemy. This question is equivalent of being in the final level of Super Mario facing Bowser)  Please write code (Python or Java) to query and print to screen all Crunchbase company “name”s and “twitter_username” (and “_id” but not the entire document) that (i) were founded between 2000 and 2010 (including 2000 and 2010), or (ii) email address is ending in “@gmail.com”.  (25% of points)

```python
#print to screen all Crunchbase company “name”s and “twitter_username” and “_id” that (i) were founded between 2000 and 2010 (including 2000 and 2010), or (ii) email address is ending in “@gmail.com”
cursor = collect_crun.find({"$or": [{"founded_year": {"$gte": 2000, "$lte": 2010}}, {"email_address": {"$regex": ".+?@gmail\.com"}}]},{"name":1, "twitter_username":1})
for i in cursor:
    print(i)
```
